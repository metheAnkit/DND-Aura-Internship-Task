**Tuple Operations**

**1. Create a tuple with 5 elements**

```python
t = (10, 20, 30, 40, 50)
print(t)
```

**2. Access the 3rd element of a tuple**

```python
t = (10, 20, 30, 40, 50)
print(t[2])
```

**3. Find the length of a tuple**

```python
t = (10, 20, 30, 40, 50)
print(len(t))
```

**4. Check whether an element exists in a tuple**

```python
t = (10, 20, 30, 40, 50)

print(30 in t)
print(60 in t)
```

**5. Convert a list into a tuple**

```python
lst = [1, 2, 3, 4]
t = tuple(lst)
print(t)
```

**6. Find maximum and minimum in a tuple**

```python
t = (5, 2, 9, 1, 7)
print(max(t))
print(min(t))
```

**7. Count how many times an element appears**

```python
t = (1, 2, 2, 3, 2, 4)
print(t.count(2))
```

**8. Find the index of an element**

```python
t = (10, 20, 30, 40)
print(t.index(30))
```

**9. Reverse a tuple**

```python
t = (1, 2, 3, 4)
rev = t[::-1]
print(rev)
```

**10. Slice tuple to get first 3 elements**

```python
t = (10, 20, 30, 40, 50)
print(t[:3])
```
