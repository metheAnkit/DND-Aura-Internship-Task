**Set Operations**

**1. Create a set with 5 elements**

```python
s = {10, 20, 30, 40, 50}
print(s)
```

**2. Add an element to a set**

```python
s = {1, 2, 3}
s.add(4)
print(s)
```

**3. Remove an element from a set**

```python
s = {1, 2, 3, 4}
s.remove(3)
print(s)
```

**4. Check if an element exists in a set**

```python
s = {10, 20, 30}

print(20 in s)
print(40 in s)
```

**5. Convert a list into a set**

```python
lst = [1, 2, 2, 3, 4]
s = set(lst)
print(s)
```

**6. Find the union of two sets**

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
```

**7. Find the intersection of two sets**

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
```

**8. Find the difference between two sets**

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.difference(b))
```

**9. Find the symmetric difference**

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.symmetric_difference(b))
```

**10. Check if one set is a subset of another**

```python
a = {1, 2}
b = {1, 2, 3, 4}

print(a.issubset(b))
```

**11. Check if two sets are disjoint**

```python
a = {1, 2}
b = {3, 4}

print(a.isdisjoint(b))
```

**12. Remove duplicate elements from a list**

```python
lst = [1, 2, 2, 3, 3, 4]
unique = list(set(lst))
print(unique)
```
