**1. get(key[, default])**

**Question: Get "salary" and return 0 if not present**

```python
d = {"name": "Alex", "age": 25}
print(d.get("salary", 0))
```

**2. keys()**

**Question: Print all keys**

```python
d = {"a": 1, "b": 2, "c": 3}
print(d.keys())
```

**3. values()**

**Question: Print all values**

```python
d = {"x": 10, "y": 20, "z": 30}
print(d.values())
```

**4. items()**

**Question: Print keyâ€“value pairs**

```python
d = {"name": "John", "age": 30}

for k, v in d.items():
    print(k, ":", v)
```

**5. update()**

**Question: Add city and update age**

```python
d = {"name": "Emma", "age": 28}
d.update({"city": "London", "age": 29})
print(d)
```

**6. pop(key[, default])**

**Question: Remove key "b" and store its value**

```python
d = {"a": 1, "b": 2, "c": 3}
x = d.pop("b")
print(x)
print(d)
```

**7. popitem()**

**Question: Remove last inserted item**

```python
d = {"a": 1, "b": 2, "c": 3}
item = d.popitem()
print(item)
print(d)
```

**8. clear()**

**Question: Remove all items**

```python
d = {"x": 1, "y": 2}
d.clear()
print(d)
```

**9. copy()**

**Question: Create independent copy**

```python
d1 = {"a": 10, "b": 20}
d2 = d1.copy()

d2["c"] = 30
print(d1)
print(d2)
```

**10. setdefault(key[, default])**

**Question: Add "country" if not present**

```python
d = {"name": "Alice"}
d.setdefault("country", "India")
print(d)
```

**11. fromkeys(iterable, value)**

**Question: Create dictionary with keys and value 0**

```python
keys = ["a", "b", "c"]
d = dict.fromkeys(keys, 0)
print(d)
```

**12. Create a dictionary from two lists: keys = ['a', 'b', 'c'], values = [1, 2, 3].**

```python
keys = ['a', 'b', 'c']
values = [1, 2, 3]

d = dict(zip(keys, values))
print(d)
```

**13. Check if a key "salary" exists in a dictionary**

```python
d = {"name": "Alex", "age": 25}

if "salary" in d:
    print("Key exists")
else:
    print("Key does not exist")
```

**14. Find the length of a dictionary**

```python
d = {"a": 1, "b": 2, "c": 3}
print(len(d))
```

**15. Merge two dictionaries {'a': 1, 'b': 2} and {'c': 3, 'd': 4}**

```python
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}

d1.update(d2)
print(d1)
```
